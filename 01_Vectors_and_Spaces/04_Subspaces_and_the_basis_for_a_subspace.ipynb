# Subspaces and the basis for a subspace

## Linear subspaces

Subspace of $\mathbb{R}^n$. A set of vectors $V$ is some subset of $\mathbb{R}^n$.

$\mathbb{R}^n$ is an infinitely large set of vectors, where each of those vectors have $n$ components.

$$\mathbb{R}^n = \left\{ \begin{bmatrix} x_1 \\ x_2 \\ ... \\ x_n \end{bmatrix} \; x_i \in \mathbb{R} \; 1 \leq i \leq n \right\}$$

$V$ is a subspace of $\mathbb{R}^n$ means: 
1. $V$ contains $\vec{0}$
2. If $\vec{x} \in V$, then $c\vec{x} \in V$ where $c \in \mathbb{R}$
    * This is closure under scalar multiplication. 
    * Explanation: "If I take some member of my set and I multiply it by some scalar, I'm still going to be in my set. If I would end up with some vector that's not in my subset, then this wouldn't be a subspace.
3. If $\vec{a}, \vec{b} \in V$, then $\vec{a} + \vec{b} \in V$
    * This is closure under addition.
    

    
### Example 1
$V = \vec{0}$ is a subspace of $\mathbb{R}^n$


### Example 2
Given: Set of vectors $S = \left\{ \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \in \mathbb{R}^2 | x_1 \geq 0 \right\}$

Find: Is $S$ a subspace of $\mathbb{R}^2$?
1. contains $\begin{bmatrix} 0 \\ 0 \end{bmatrix}$
2. it is closed under addition
3. it is not closed under scalar multiplication

Conclusion: $S$ is not a subspace of $\mathbb{R}^2$.

### Example 3
Given: Span of set of vectors $U = span(v_1, v_2, v_3$

Find: Is $U$ a valid subspace of $\mathbb{R}^n$
1. $0\vec{v}_1 + 0\vec{v}_2 + 0\vec{v}_3 = \vec{0}$
2. In order to be a member of the set, a vector can be represented as a linear combination of the vectors in the span.
$$\vec{x} = c_1 \vec{v}_1 + c_2 \vec{v}_2 + c_3  \vec{v}_3$$

$$\begin{equation} \begin{split} a\vec{x} 
&= ac_1 \vec{v}_1 + ac_2 \vec{v}_2 + ac_3  \vec{v}_3 \\
&= c_4 \vec{v}_1 + c_5 \vec{v}_2 + c_6  \vec{v}_3 \end{split} \end{equation}$$
    *  $U$ is closed under multiplication.
3. $$\vec{y} = d_1 \vec{v}_1 + d_2 \vec{v}_2 + d_3  \vec{v}_3$$
$$\begin{equation} \begin{split} \vec{x} + \vec{y} 
&= (c_1 + d+1) \vec{v}_1 + (c_2 + d_2) \vec{v}_2 + (c_3 + d_3)  \vec{v}_3 \\
&= d_4 \vec{v}_1 + d_5 \vec{v}_2 + d_6  \vec{v}_3 
\end{split} \end{equation}$$
    * $U$ is closed under addition

Conclusion: $U$ is a subspace of $\mathbb{R}^n$.


## Basis of a subspace
Given: Subspace $V = span(\vec{v}_1, \vec{v}_2, ..., \vec{v}_n)$
* Note: As shown above, the span of any set of vectors is a subspace.
* All the vectors are linearly independent: You can't represent any of these vectors as a combination of any of the other vectors
* Span meant: The set/subspace represents all the possible linear combinations of all the vectors: $c_1 \vec{v}_1 + c_2 \vec{v}_2 + ... + c_n \vec{v}_n \; c \in \mathbb{R}$

The set of vectors $S = \{\vec{v}_1, \vec{v}_2, ..., \vec{v}_n\}$ is a basis for $V$.
* If something is a basis for a subspace, that means that if you take the span of the vectors, you can construct any of the vectors in the subspace, and the vectors are linearly independent.
* A basis is a minimum set of vectors that spans the subspace

**Standard basis** for $\mathbb{R}^2$:

$$T = \left\{ \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \, \begin{bmatrix} 0 \\ 1 \end{bmatrix} \right\}$$

Claim: Using a basis, one can represent any vector in the subspace by some unique combination of the vectors in your basis. 

Proof: Given: Set $\{v_1, v_2, ..., v_n\} =$ Basis for some subspace $U$. That means the vectors are linearly independent, and that the span of them will give me all the possible vectors (also called members or components) of $U$.

Each member of $U$ can only be defined by a unique set of combinations of the vectors.

Let's say $\vec{a} \in U$, that means that $\vec{a}$ can be represented by a linear combination of the vectors: $\vec{a} = c_1 \vec{v}_1 + c_2 \vec{v}_2 + ... + c_n \vec{v}_n$.

Let's do a proof by contradiction and assume that $\vec{a}$ can be represented by another linear combination: $\vec{a} = d_1 \vec{v}_1 + d_2 \vec{v}_2 + ... + d_n \vec{v}_n$.

By subtracting the two equations, we get: $\vec{0} = (c_1 - d_1) \vec{v}_1 + (c_2 - d_2) \vec{v}_2 + ... + (c_n - d_n) \vec{v}_n$

The only solution this equation has is if all constants in front of the vectors are $0$ as per definition of linear independence, meaning that $c_1 = d_1$, $c_2 = d_2$, and so on. 

So, if you have a basis for some subspace, any member of that subspace can be uniquely determined by a unique combination of the vectors.
