# Linear dependence and independence

## Introduction to linear independence
Given:
$$\left\{ 
\begin{bmatrix} 2 \\ 3 \end{bmatrix} , \; 
\begin{bmatrix} 4 \\ 6 \end{bmatrix}  \right\}$$

What is the span of these vectors? What are all of the vectors that these two vectors can represent?

Span: All the vectors that can be represented by linear combinations of these. 

$$ \begin{equation} \begin{split}
span\left( \left\{ \begin{bmatrix} 2 \\ 3 \end{bmatrix} , \; 
\begin{bmatrix} 4 \\ 6 \end{bmatrix} \right\} \right) &= 
c_1 \cdot \begin{bmatrix} 2 \\ 3 \end{bmatrix} + c_2 \cdot
\begin{bmatrix} 4 \\ 6 \end{bmatrix} \\
&= (c_1 + 2 c_2) \cdot \begin{bmatrix} 2 \\ 3 \end{bmatrix} \\
&= c_3 \cdot \begin{bmatrix} 2 \\ 3 \end{bmatrix}
\end{split} \end{equation}$$

Instead of talking about linear combinations of two vectors, I can reduce this to just a scalar combination of one vector.

If you have two collinear vectors in $\mathbb{R}^2$, they reduce to just one line.

We call the given set linearly dependent. This means that one of the vectors in the set can be represented by some combination of the other vectors in the set.



Given two vectors that span $\mathbb{R}^2$, a third vector can't be coplanar with the two vectors if we want to span $\mathbb{R}^3$. If it was coplanar, it's not "adding any more directionality".

If there are three vectors that span $\mathbb{R}^3$, then they are linearly independent.

Given:
$$\begin{bmatrix} 2 \\3 \end{bmatrix} 
\begin{bmatrix} 7 \\2 \end{bmatrix}
\begin{bmatrix} 9 \\5 \end{bmatrix}$$

$$\vec{v}_1 + \vec{v}_2 = \vec{v}_3$$

meaning $\vec{v}_3$ is a linear combination of $\vec{v}_1$ and $\vec{v}_2$. So, $\{\vec{v}_1, \vec{v}_2, \vec{v}_3\}$ is a linearly dependent set. $span(\vec{v}_1, \vec{v}_2, \vec{v}_2) \in \mathbb{R}^3$

It is a general idea that if you have three 2-dimensional vectors, one of them is going to be redundant. That's because $span(\vec{v}_1, \vec{v}_2) \in \mathbb{R}^2$. 

## More in linear independence
$s = \{v_1, v_2, ..., v_n\}$ are linearly dependent iff ($\iff$) I can find a set of constants $c_1,...,c_n$ such that $c_1 v_1 + c_2 v_2 + ... + c_n v_n = 0$ where not all are zero.

Linear dependence is if I can represent one of the vectors as a combination of the other vectors. For example

$$
\begin{equation}
\begin{split}
v_1 &= a_2 v_2 + a_3 v_3 + ... + a_n v_n \\
0 &= -1 \cdot v_1 + a_2 v_2 + a_3 v_3 + ... + a_n v_n
\end{split}
\end{equation}
$$

Which satisfies the condition for linear dependence, since $c_1 = -1$. This proofs that both definitions of linear dependence are equivalent.

## Span and linear independence example
Given:
$$\left\{ 
\begin{bmatrix} 1 \\ -1 \\ 2 \end{bmatrix},
\begin{bmatrix} 2 \\  1 \\ 3 \end{bmatrix},
\begin{bmatrix} -1\\  0 \\ 2 \end{bmatrix}
\right\} = S$$

Is $span(S) = \mathbb{R}^3$? I.e. is $S$ linearly independent?

It is if I can represent any vector $\vec{v} \in \mathbb{R}^3$ with a linear combination of the three vectors, i.e.:

$$
c_1 \begin{bmatrix} 1 \\ -1 \\ 2 \end{bmatrix} + 
c_2 \begin{bmatrix} 2 \\  1 \\ 3 \end{bmatrix} + 
c_3 \begin{bmatrix} -1\\  0 \\ 2 \end{bmatrix} =
\begin{bmatrix} a\\  b \\ c \end{bmatrix}
$$

Solving for $c_1$, $c_2$, and $c_3$ yields equations that are valid for any $a$, $b$, $c$. In order for the linear combination to equal $\vec{0}$, all constants $c_1$, $c_2$, and $c_3$ have to be zero, meaning that the set of vectors is linearly independent.

If I have exactly three vectors and they do span $\mathbb{R}^3$, they have to be linearly independent. If they weren't linearly independent, then one of these would be redundant.
