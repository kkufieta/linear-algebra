# Vectors

## Vector intro for linear algebra

* Vector: magnitude + direction
  * e.g. 5 mph (speed, which is a scalar) + east == velocity
  
Note:
* Speed is a scalar
* Velocity is a vector
* Vectors are independent of location. A vector starting at (0, 0) is equivalent to a vector starting at (1, 1) as long as they have the same magnitute + direction.



## Real coordinate spaces
Notation:
$$\vec{v} = \begin{bmatrix} 5 \\ 0 \end{bmatrix} \in \mathbb{R}^2$$

where $\mathbb{R}^2$ is the 2-dimensional real coordinate space.

The 2-dimensional real coordinate space is all the possible real-valued 2-tuples.

A tuple is an ordered list of numbers.

### Exampe of 3-tuples
$\mathbb{R}^3$: 3-dimensional real coordinate space, described by real-valued 3-tuples

$$\vec{x} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix} \in \mathbb{R}^3$$

$$\vec{b} = \begin{bmatrix} 3 \\ -2 \\ 10 \end{bmatrix} \in \mathbb{R}^3$$



## Adding vectors algebraically & graphically

$$\vec{a} = \begin{bmatrix} 6 \\ -2 \end{bmatrix}, \; \vec{b} = \begin{bmatrix} -4 \\ 4 \end{bmatrix}, \; \vec{a}, \vec{b} \in \mathbb{R}^2$$

$$\vec{a} + \vec{b} = \vec{b} + \vec{a} = \begin{bmatrix} 2 \\ 2 \end{bmatrix}$$ 

**Note:** Vectors don't only apply to displacement. They can apply to velocity and acceleration, and a whole series of things.

```python
import numpy as np
import matplotlib.pyplot as plt

V = np.array([[6, -2], [-4, 4], [2, 2]])
origin = np.array([[0, 6, 0],[0, -2, 0]]) # origin point

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-1, 7)
plt.ylim(-3, 3)
```

## Multiplying a vector by a scalar
$$\vec{a} = \begin{bmatrix} 2 \\ 1 \end{bmatrix} $$

$$3\vec{a} = 3\begin{bmatrix} 2 \\ 1 \end{bmatrix} = \begin{bmatrix} 3 * 2 \\ 3 * 1 \end{bmatrix} = \begin{bmatrix} 6 \\ 3 \end{bmatrix} $$

```python
import numpy as np
import matplotlib.pyplot as plt

V = np.array([[2, 1], [6, 3], [-1, -0.5]])
origin = np.array([[0, 0, 0],[0, 0, 0]]) # origin point

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-2, 7)
plt.ylim(-1, 5)
```

## Unit vectors intro
Denote a unit vector: Use a hat instead of an arrow: $\hat{i}$

Given the unit vectors:
$$ \hat{i} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \; \hat{j} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$$

we can represent any 2 dimensional vector as a sum of scaled up versions of $\hat{i}$ and $\hat{j}$.

For example:
$$ \vec{v} = \begin{bmatrix} 2 \\ 3 \end{bmatrix} = 2 \cdot \hat{i} + 3 \cdot \hat{j}$$

Once you get to three dimensions, you introduce $\hat{k} = \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}$.

## Parametric representation of lines
Set of colinear vectors:
$$S = \{c \cdot \vec{v} \,|\, c \in \mathbb{R}\}$$ given $$\vec{v} = \begin{bmatrix} 2 \\ 1 \end{bmatrix}$$

If we view these vectors as position vectors (meaning a vector represents a point in space $\mathbb{R}^2$, the set is going to be represented as a line. You can think of $\vec{v}$ as a "slope vector". 

Next, given $\vec{x} = [2, 4]^T$, we define a new set of vectors $L = \{\vec{x} + t\cdot \vec{v} \,|\, t \in \mathbb{R} \}$, which is also represented as a line in 2D. 



### Why bother thinking like this about lines?
This is a limited line equation that works well in $\mathbb{R}^2$:
$$y = mx + b$$

Whereas the new definition using a set is much more general:
$$L = \{\vec{x} + t\cdot \vec{v} \,|\, t \in \mathbb{R} \}$$
where $\vec{x}$ and $\vec{v}$ both $\in \mathbb{R}^2$, meaning they're 2-dimensional vectors, but we can extend it to an arbitrary number of dimensions. 

In the classic algebra problem where you need to find the equation for the line (here: the set definition for the line). 

### Find the parameterization of a line in 2D
Given two vectors / points $\vec{a} = [2, 1]^T$ and $\vec{b} = [0, 3]^T$, find the parameterization of the line that goes through these two points. 

Two possible solutions: 
$$ L = \{\vec{b} + t(\vec{b} - \vec{a}) | t \in \mathbb{R} \}$$
$$ L = \{\vec{a} + t(\vec{b} - \vec{a}) | t \in \mathbb{R} \}$$

Using the vectors given above, we get:
$$ L = \{\begin{bmatrix} 0 \\ 3 \end{bmatrix} + t\begin{bmatrix} -2 \\ 2 \end{bmatrix} | t \in \mathbb{R} \}$$

Written as parametric equations:
$$ \vec{L} = \begin{bmatrix} l_1 \\ l_2 \end{bmatrix} = \begin{bmatrix} -2t \\ 2t + 3 \end{bmatrix}$$

The traditional parametric definition of the line is:
$$x = -2t$$
$$y = 2t + 3$$

### Find the parameterization of a line in 3D
In $\mathbb{R}^3$ the only way to define a line is to have a parametric equation.

Given two three-dimensional points $P_1$ and $P_2$, the line going through them can be defined as
$$ L = \{\vec{P}_1 + t \cdot (\vec{P}_1 - \vec{P}_2) \;|\; t \in \mathbb{R} \}$$

E.g. $x + y + z = k$ is a plane, not a line.

## Add vectors: magnitude & direction to component
[Review topic on Khan Academy](https://www.khanacademy.org/math/precalculus/x9e81a4f98389efdf:vectors/x9e81a4f98389efdf:component-form/a/vector-magnitude-and-direction-review)

### Problem 1
Given:
* $\vec{v}$ with magnitude $8$ and direction angle $140$
* $\vec{w}$ with magnitude $4$ and direction angle $40$

Find:
* $\vec{v} + \vec{w}$

$$\vec{v} + \vec{w} = 
\begin{pmatrix} 
8 \cdot cos(140) + 4 \cdot cos(40)  \\
8 \cdot sin(140) + 4 \cdot sin(40) 
\end{pmatrix} = 
\begin{pmatrix} 
-3.064 \\
7.7134 
\end{pmatrix} $$

```python
import numpy as np
import math

r = np.array([8, 4])
theta = np.array([math.radians(140), math.radians(40)])
result = [sum(r * np.cos(theta)), sum(r * np.sin(theta))]
print("v + w = ", result)
```

### Problem 2
Given: Vectors in form $(r, \theta)$ where $r$ is the magnitude and $\theta$ is the direction angle
* $\vec{v} = (7, 160)$
* $\vec{w} = (9, 20)$

Find:
* $\vec{v} + \vec{w}$

```python
r = np.array([7, 9])
theta = np.array([math.radians(160), math.radians(20)])
result = [sum(r * np.cos(theta)), sum(r * np.sin(theta))]
print("v + w = ", result)
```
